```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv('cost_revenue_clean.csv')

data.describe()
```

```python
x = pd.DataFrame(data, columns= ['production_budget_usd'])
y = pd.DataFrame(data, columns= ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(x, y, alpha=0.3) # transparency
plt.title("Film Cost v Global Revenue")
plt.xlabel('Production Budget in Dollars')
plt.ylabel('Worldwide Gross in Dollars')

plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```

## Comuting the Line


Residual Sum of Squares

```python
regression = LinearRegression()
regression.fit(x, y)
```

Slope Coefficient:

```python
regression.coef_ # theta_1
```

```python
# Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(x, y, alpha=0.3) # transparency
plt.plot(x, regression.predict(x), color='red', linewidth='4')
plt.title("Film Cost v Global Revenue")
plt.xlabel('Production Budget in Dollars')
plt.ylabel('Worldwide Gross in Dollars')

plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```

The goodness of fit
R^2 

```python
# amount of variation in film revenue that is explained by the film budget
regression.score(x, y)
```
