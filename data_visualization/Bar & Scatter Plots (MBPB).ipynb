```python
def get_reviews():
    return pd.read_csv("fandango_scores.csv")
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from numpy import arange

norm_reviews = get_reviews()
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

# get the values from the reviews 
bar_heights = norm_reviews[num_cols].iloc[0].values

# generate evenly spaced values from an int
bar_positions = arange(5) + 0.75

# create the figure and the axes
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)

plt.plot()
plt.show()
```

## Plotting Bar Charts with Labels

```python
norm_reviews = get_reviews()

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75

tick_positions = range(1,6)

fig, ax = plt.subplots()

ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

## Plotting Horizontal Bar Charts

```python
norm_reviews = get_reviews() 

import matplotlib.pyplot as plt
from numpy import arange

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_widths = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
tick_positions = range(1,6)

fig, ax = plt.subplots()

ax.barh(bar_positions, bar_widths, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

## Scatter Plots

Correlation Coefficient - A correlation coefficient is a numerical measure of some type of correlation, meaning a statistical relationship between two variables
https://en.wikipedia.org/wiki/Correlation_coefficient

```python
norm_reviews = get_reviews() 

fig, ax = plt.subplots()
fandango = norm_reviews["Fandango_Ratingvalue"]
rotten_tomatoes = norm_reviews["RT_user_norm"]

ax.scatter(fandango, rotten_tomatoes)
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")
ax.set_title("Correlation between Fandango and Rotten Tomato reviews")
plt.show()
```
