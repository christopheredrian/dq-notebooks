To help you become familiar with matplotlib, we'll focus on the first 2 use cases. When working with commonly used plots in matplotlib, the general workflow is:

    create a plot using data
    customize the appearance of the plot
    display the plot
    edit and repeat until satisfied

````python
import pandas as pd 
import matplotlib.pyplot as plt 

# read the csv file 
unrate = pd.read_csv('unrate.csv')

# convert the date column from string to date64 type
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

# get only the first 12 for bot x and y axis 
x_axis = unrate["DATE"].head(12)
y_axis = unrate["VALUE"].head(12)

# rotate the x-axis labels to 90 deg
plt.xticks(rotation=90)

# set the x axis label to "Month"
plt.xlabel("Month")

# set the y axis label to "Unemployment Rate"
plt.ylabel("Unemployment Rate")

# set the title of the Graph
plt.title("Monthly Unemployment Trends, 1948")

# plot the 
plt.plot(x_axis, y_axis)

# show on screen
plt.show()
````

- How to plot multiple subplots
```python3
import matplotlib.pyplot as plt
fig = plt.figure()

ax2 = fig.add_subplot(2,1,2)

```

````python
import matplotlib.pyplot as plt

fig = plt.figure()

# creates 2x2 plot 

# add first plot at the top left quadrant
ax1 = fig.add_subplot(2,2,1)

# add another plot at the bottom right quadrant
ax2 = fig.add_subplot(2,2,4)

plt.show()
````
